### **Anteriormente 3**
Ha llegado el momento de poner a prueba los conocimientos adquiridos durante la lección. Continuando con el proyecto de las naranjas/toronjas, ahora debes calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$


La solución para la pendiente es:

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

Y para la intersección:
 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

Y obtenemos la ecuación de la recta:

$$y = \alpha x + \beta $$

```python
# Naranjas
n = np.size(naranjas_weight)
x = naranjas_diameter
y = naranjas_weight

m = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) \
    / (n * np.sum(x ** 2) - np.sum(x) ** 2)
b = np.mean(y) - m * np.mean(x)
print(f"La pendiente es: {m}\nLa intersección es: {b}")

y = m * x + b
print(f"y is: {y}")

print(f"Matrix norm: {np.linalg.norm(naranjas_weight - y)}")

# Si queremos definir este conjunto de datos con una recta esta es la
# correcta
plt.plot(naranjas_diameter, naranjas_weight)
plt.plot(x, y)
plt.show()
```

```python
# Toronjas
n = np.size(toronjas_weight)
x = toronjas_diameter
y = toronjas_weight


m = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) \
    / (n * np.sum(x ** 2) - np.sum(x) ** 2)
b = np.mean(y) - m * np.mean(x)
print(f"La pendiente es: {m}\nLa intersección es: {b}")

y = m * x + b
print(f"y is: {y}")

print(f"Matrix norm: {np.linalg.norm(toronjas_weight - y)}")

# Si queremos definir este conjunto de datos con una recta esta es la
# correcta
plt.plot(toronjas_diameter, toronjas_weight)
plt.plot(x, y)
plt.show()
```
